```python
import pandas as pd
import numpy as np
import matplotlib as plt
import copy, math

std_data = pd.read_csv('Student_Performance.csv')
std_data.columns = std_data.columns.str.replace(' ', '_')
std_data
#print(type(std_data))
#print(std_data.info())
```

```python
#std_data.rename(columns={'Extracurricular Activities' : 'Extracurricular'}, inplace = True)
#std_data
std_data = std_data.replace('Yes', '1', regex = True)
std_data = std_data.replace('No', '0', regex = True)
std_data['Extracurricular_Activities'] = std_data['Extracurricular_Activities'].astype('int64')
std_data
```

```python
def normalise_data(arr):
  #min_val = np.min(arr)
  #max_val = np.max(arr)
  mean = np.mean(arr)
  std = np.std(arr)
  norm_arr = np.zeros(len(arr))
  for i in range(len(arr)):
    norm_arr[i] = (arr[i] - mean) / std
  return norm_arr
```

```python
for i in range(std_data.shape[1]-1):
  std_data.iloc[:, i] = normalise_data(std_data.iloc[:, i])
std_data
```

```python
number = std_data.shape[0]
print(number)
number_train = int(number * 9 / 10)
```

```python
x_train_list = []
for j in range(std_data.shape[1]):
    xj_train = np.zeros(number_train)
    for i in range(number_train):
        xj_train[i] = std_data.iloc[i, j]
    x_train_list.append(xj_train)  # Append xj_train directly to the list

# Separate the features (x_train) from the target (y_train)
x_train_tup = tuple(x_train_list[:-1]) # Exclude the last column (target)
x_train = np.column_stack(x_train_tup)
y_train = x_train_list[-1]  # The last column is the target

print(x_train)
print(y_train)
```

```python
w_para = np.zeros(std_data.shape[1] - 1)
print(w_para)
b_para = 0
print(b_para)
```

```python
def prediction(x, w, b):
  c = np.dot(x, w) + b
  return c
```

```python
def predict():
  predic= np.zeros(number_train)
  for i in range(number_train):
    predic[i] = prediction(x_train[i], w_para, b_para)
    #print(predic)
  return predic
```

```python
pred = predict()
```

```python
def loss_calc(pred, y):
  loss = 1/2*((pred - y) ** 2)
  return loss
```

```python
def lossss():
  losses = np.zeros(number_train)
  for i in range(number_train):
    losses[i] = loss_calc(pred[i], y_train[i])
  #print(losses)
  return losses
```

```python
loss = lossss()
```

```python
def cost_function(loss, l, w):
  costie = np.mean(loss) + (l * np.sum(np.square(w)))/(2*len(loss))
  #print(costie)
  return costie
```

```python
cost = cost_function(loss, 1, w_para)
```

```python
def grad_desc(w, b, cost, pred, x, y, lam, alp, n):
  temp_w = copy.deepcopy(w)
  dj_dw = np.zeros(len(w))
  for k in range(len(w)):
    for i in range(number_train):
      dj_dw[k] = (((pred[i] - y[i])/n)*x[i][k]) + ((lam*w[k])/n)
      temp_w[k] = w[k] - (alp * dj_dw[k])
      dj_db = (pred[i] - y[i])/n
      temp_b = b - (alp * dj_db)
    w[k] = temp_w[k]
    b = temp_b
  return w, b
  #print(w_para)
  #print(b_para)
```

```python
w_para, b_para = grad_desc(w_para, b_para, cost, pred, x_train, y_train, 0.1, 1, number_train)
```

```python
i = 0
while cost > 0.1:
  pred = predict()
  loss = lossss()
  cost = cost_function(loss, 1, w_para)
  w_para, b_para = grad_desc(w_para, b_para, cost, pred, x_train, y_train, 0.1, 1, number_train)
  print(f'{i}th iteration, {cost}')
  i += 1
else:
  print('done')
  print(w_para)
  print(b_para)
  print(cost)
  print(f'done in {i} iterations')
```
