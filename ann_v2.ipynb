```python
def one_hot_label(arr, unique):
  arrofarr_list = []
  for j in range(len(unique)):
    arr_name = np.empty(len(arr), dtype = object)
    for i in range(len(arr)):
      arr_name[i] = arr[i]
    arrofarr_list.append(arr_name)

  arrofarr_tup = tuple(arrofarr_list)
  arrofarr = np.row_stack(arrofarr_tup)
  for j in range(len(unique)):
    for i in range(len(arr)):
      if arrofarr[j][i] == unique[j]:
        arrofarr[j][i] = '1'
      else:
        arrofarr[j][i] = '0'
  #print(arrofarr)
  return arrofarr
```

```python
name_arr, unique_names = extraction(0)
```

```python
name_arrofarr = one_hot_label(name_arr, unique_names)
print(name_arrofarr)
```
